```python
# Cell 1: Pick 5 good triads and save a showcase CSV (Notebook-friendly)

import pandas as pd

input_csv = "/Users/neda/Desktop/UBC/PHD/LLM_research/agent-choice-lab/tasks/product_triads-decoy-pricerating-sequential.csv"
output_csv = "/Users/neda/Desktop/UBC/PHD/LLM_research/agent-choice-lab/tasks/triads_showcase5.csv"

df = pd.read_csv(input_csv)

# (category, keyword in target_name)
pick_specs = [
    ("Chocolate", "Demet's Original Chocolate Turtles"),
    ("Mouthwash", "The Natural Dentist Whitening"),
    ("Cookies", "Oreo Partnership Family Size"),
    ("Deodorants & Antiperspirants", "Degree Men Dry Protection"),
    ("Earbud Headphones", "Tiksounds Wireless Earbuds"),
]

rows = []
for cat, kw in pick_specs:
    r = df[(df["category"] == cat) & (df["target_name"].str.contains(kw, case=False, na=False))].head(1)
    if len(r):
        rows.append(r)

showcase = pd.concat(rows).copy()

# Resolve placeholder URLs to your local shopping env
for c in ["target_url", "competitor_url", "decoy_url"]:
    showcase[c] = showcase[c].str.replace("${env.abxlab_url}", "http://localhost:7770/", regex=False)

showcase.to_csv(output_csv, index=False)

print(f"Saved: {output_csv}")
showcase[[
    "category",
    "target_name", "target_url",
    "competitor_name", "competitor_url",
    "decoy_name", "decoy_url",
    "tc_price_premium_pct", "tc_rating_adv", "dt_price_premium_pct", "td_rating_adv"
]]
```

```python
# Cell 2: Open all URLs in browser tabs (requires local site running)

import webbrowser

for _, row in showcase.iterrows():
    for u in [row["target_url"], row["competitor_url"], row["decoy_url"]]:
        webbrowser.open_new_tab(u)

print("Opened tabs for 5 triads (15 product pages).")
```
